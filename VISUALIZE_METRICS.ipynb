```python
date_result = "2023-06-22"
# date_result = "2023-06-20"

MAIN_CODE_PATH = os.getcwd()
RESULT_PATH = MAIN_CODE_PATH + '/results/' + date_result

evagan_list_file_path = RESULT_PATH + "/{dataset}_{timestamp}/EVAGAN_CC_LISTS.csv"
classifier_list_path = RESULT_PATH + "/{dataset}_{timestamp}/CLASSIFIER_LISTS.csv"


MAIN_CODE_PATH = os.getcwd()
RESULT_PATH = MAIN_CODE_PATH + '/results/' + date_result

evagan_list_file_path = RESULT_PATH + "/{dataset}_{timestamp}/EVAGAN_CC_LISTS.csv"
classifier_list_path = RESULT_PATH + "/{dataset}_{timestamp}/CLASSIFIER_LISTS.csv"

mapping = {"EVAGAN": {"CIC-2018": "04_28_14", "EdgeIIoT": "01_30_34", "InSDN": "04_58_10"},
           "Transformer-EVAGAN": {"CIC-2018": "18_58_29", "EdgeIIoT": "01_50_21", "InSDN": "18_58_00"}}


mapping = {"Transformer-EVAGAN-IDS": {"CIC-2018": "05_00_01", "EdgeIIoT": "09_44_18", "InSDN": "09_07_57"},
           "Fed-Transformer-EVAGAN-IDS": {"CIC-2018": "23_59_35", "EdgeIIoT": "01_39_44", "InSDN": "11_33_13"}}

# mapping = {"Fed-Transformer-EVAGAN-IDS": {"CIC-2018": "23_59_35", "EdgeIIoT": "01_39_44", "InSDN": "11_33_13"}}
```

```python
for model, datasets in mapping.items():
    for dataset, timestamp in datasets.items():
        full_evagan_list_file_path = evagan_list_file_path.format(dataset=dataset, timestamp=timestamp)

        with open(full_evagan_list_file_path, 'r') as file:
            print(full_evagan_list_file_path)
            reader = csv.reader(file)
            data = list(reader)

        epochs = [int(row[0]) for row in data[1:]]
        evaluation1 = [float(row[1]) for row in data[1:]]
        evaluation2 = [float(row[2]) for row in data[1:]]
        evaluation3 = [float(row[3]) for row in data[1:]]
        evaluation4 = [float(row[4]) for row in data[1:]]

        # Tạo biểu đồ
        plt.plot(epochs, evaluation1, label='GEN_VAL')
        plt.plot(epochs, evaluation2, label='FAKE_BOT')
        plt.plot(epochs, evaluation3, label='REAL_NOR')
        plt.plot(epochs, evaluation4, label='REAL_BOT')
        plt.xlabel('Epochs')
        plt.ylabel('Evaluation Scores')
        plt.legend()
        plt.savefig(RESULT_PATH + "/{dataset}_{timestamp}/EVAGAN_Estimations.png".format(dataset=dataset, timestamp=timestamp))
        plt.show()
```

```python
color_dict = {
    'D_Loss_Real_Bot': 'gold',
    'D_Loss_Fake_Bot': 'mediumseagreen',
    'D_Loss_Real_Normal': 'mediumpurple',
    'G_Loss': 'dodgerblue'
}

for model, datasets in mapping.items():
    for dataset, timestamp in datasets.items():
        full_evagan_list_file_path = evagan_list_file_path.format(dataset=dataset, timestamp=timestamp)

        with open(full_evagan_list_file_path, 'r') as file:
            print(full_evagan_list_file_path)
            reader = csv.reader(file)
            data = list(reader)

        epochs = [int(row[0]) for row in data[1:]]
        loss1 = [float(row[5]) for row in data[1:]]
        loss2 = [float(row[6]) for row in data[1:]]
        loss3 = [float(row[7]) for row in data[1:]]
        loss4 = [float(row[9]) for row in data[1:]]

        # Tạo biểu đồ
        plt.plot(epochs, loss1, label='D_Loss_Real_Bot', color=color_dict['D_Loss_Real_Bot'])
        plt.plot(epochs, loss2, label='D_Loss_Fake_Bot', color=color_dict['D_Loss_Fake_Bot'])
        plt.plot(epochs, loss3, label='D_Loss_Real_Normal', color=color_dict['D_Loss_Real_Normal'])
        plt.plot(epochs, loss4, label='G_Loss', color=color_dict['G_Loss'])

        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        # plt.savefig(RESULT_PATH + "/{dataset}_{timestamp}/EVAGAN_Loss.png".format(dataset=dataset, timestamp=timestamp))
        plt.show()
```
